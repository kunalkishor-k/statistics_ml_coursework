### Task 6: mean values with 95% confidence intervals for Y(weight) distributions.
Assuming that the data are a sample from a larger population.

```python
import statistics as st
```

```python
sample_mean_weight = st.mean(df['weight'])
sample_mean_weight
```

```python
sigma_weight = st.stdev(df['weight'])
sigma_weight
```

```python
# confidene Interval
upper_bound = sample_mean_weight+2*sigma_weight
lower_bound = sample_mean_weight-2*sigma_weight
print('confidence_interval')
print('lower_bound:',lower_bound)
print('upper_bound:',upper_bound)
```

### Task 7: mean values with 95% confidence intervals for X(catch_time) distributions.

```python
sample_mean_catch_time = st.mean(df['catch_time'])
sample_mean_catch_time
```

```python
sigma_catch_time = st.stdev(df['catch_time'])
sigma_catch_time
```

```python
# confidene Interval
upper_bound = sample_mean_catch_time+2*sigma_catch_time
lower_bound = sigma_catch_time-2*sigma_catch_time
print('confidence_interval')
print('lower_bound:',lower_bound)
print('upper_bound:',upper_bound)
```
